```python
import numpy as np
import matplotlib.pyplot as plt
from ipywidgets import interact
import math
%matplotlib inline
```

# . Loss funtion為 $L(x)=x^4 + x^3 + 9x^2sin(5) + 5$ 
# . 微分侯的損失函數為  $l(x)=4x^3 + 3x^2 + 18sin(5)x$ 

```python
#先定義我要用的損失函數及它的微分結果
def Loss(x):
    return x**4 + x**3 + (9*x**2)*np.sin(5) + 5
def DLoss(x):
    return 4*x**3 + 3*x**2 + 18*np.sin(5)*x
```

```python
#設定X來花出來的函數的範圍，y寫我要畫出來的損失函數
x = np.linspace(-4,3, 100)
y = x**4 + x**3 + (9*x**2)*np.sin(5) + 5
```

```python
#寫出找Miminum的Code!
def Get_Min(start, IR, iteration):
    #start = np.random.uniform(-2, 2)
    plt.figure(figsize=(10,5))
    plt.plot(x,y)
    plt.scatter(start,Loss(start),color = 'r')
    plt.xlabel("x", size = "19")
    plt.ylabel("y", size = "19")
    plt.title("Loss(x)", size = "19")
    plt.annotate('  Starting Point', xy = (3, 33), arrowprops={'color' : 'blue'})
    plt.annotate('    Local Mimimun', xy = (1.8, -8), arrowprops={'color' : 'blue'})
    plt.annotate('  Global Mimimun', xy = (-2.5, -27), arrowprops={'color' : 'blue'})
    #n : learning rate
    g=start
    n = IR
    i = iteration
    for i in range(i):
        c = i+1
        g = g - n* DLoss(g)
        plt.scatter(g,Loss(g), s = 20, alpha = 1-1/c, c ="g")
    
```

# 搭配 Interact 功能畫出來看看

```python
interact(Get_Min, start=(-4,3),IR=(0.005, 0.1),iteration=(1,30))
```

```python
 Get_Min(3, 0.005, 15)
```

 * 這結果表示 進去Local Minimum, 所以以下用不同的方法實驗達到 Global Minimum
     1. 改變Learning Rate 跟 Iteration看看
     2. 改Starting Point

### 改變Learning Rate 跟 Iteration看看

```python
 Get_Min(3, 0.04, 30)
```

*成功達到Global Minimum [可以知道說 若用更大的Learning Rate可以比較容易跳到我要的Convex裏面]

### 改變Learning Rate 跟 Iteration看看

```python
 Get_Min(-1, 0.005, 15)
```

*還沒達到Global Minimum就結束。 所以增加 Iteration來再次實驗

```python
 Get_Min(-1, 0.005, 40)
```

*成功達到Global Minimum

**心得 ： 找到正確的Global Minimum，需要好好調LearningRate,Iteration,而且 出發點帶來的影響也大。 這些因素會影響整個系統跑的速度**
